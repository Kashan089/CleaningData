```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from difflib import get_close_matches
```

```python
df = pd.read_csv(r'IMDB_Dataset.csv', encoding='latin1', sep=None, engine='python')
```

```python
pd.set_option('display.max_rows', None)
```

```python
df
```

```python
df = df.drop('Unnamed: 8', axis=1)
```

```python
df.columns = df.columns.str.replace(r'[^\x00-\x7F]+', '', regex=True)
```

```python
df.head(5)
```

```python
df.rename(columns={'Genr': 'Genre', 'Original titl': 'Original Title'}, inplace=True)
```

```python
df.columns = df.columns.str.strip()
```

```python
df.info()
```

```python
df['Release year'] = pd.to_datetime(df['Release year'], errors='coerce')
```

```python
df[['Original Title']]
```

```python
df['Original Title'] = (df['Original Title']
    .str.encode('latin1', errors='ignore')
    .str.decode('utf-8', errors='ignore')
    .str.strip()
)
```

```python
df[['Original Title']]
```

```python
df[['Release year']]
```

```python
df[['Duration']]
```

```python
df['Duration'] = df['Duration'].str.replace(r'[^0-9]', '', regex=True)
```

```python
df['Duration'] = df['Duration'].fillna(0).replace('', 0)
```

```python
df['Duration'] = df['Duration'].astype(int)
```

```python
df[['Duration']]
```

```python
df['Country'].value_counts(dropna=False)
```

```python
df['Country'] = df['Country'].replace({'US': 'USA'})
```

```python
df['Country'] = df['Country'].str.replace(r'[\.0-9].*$', '', regex=True)
```

```python
df['Country'] = df['Country'].astype(str)
unique_countries = df['Country'].unique()
correct_name = 'New Zealand'
for name in unique_countries:
    if get_close_matches(name, [correct_name], cutoff=0.6):
        df['Country'] = df['Country'].replace(name, correct_name)
```

```python
df['Country'] = df['Country'].replace('nan', '')
```

```python
df['Country'].value_counts(dropna=False)
```

```python
df['Content Rating'].value_counts(dropna=False)
```

```python
df['Content Rating'] = df['Content Rating'].fillna('Not Rated')
```

```python
df['Content Rating'].value_counts(dropna=False)
```

```python
df[['Director']]
```

```python
df['Director'] = df['Director'].str.encode('latin1').str.decode('utf-8', errors='ignore')
```

```python
df['Director'] = df['Director'].fillna('')
```

```python
df[['Director']]
```

```python
df[['Income']]
```

```python
df['Income'] = (df['Income']
    .astype(str)
    .str.replace('o', '0')
    .str.replace(r'[^0-9]', '', regex=True)
    .replace('', '0')
    .astype(float)
    .div(10)
    .astype(int)
)
```

```python
df[['Income']]
```

```python
df[['Votes']]
```

```python
df['Votes'] = df['Votes'].fillna(0)
```

```python
df['Votes'] = df['Votes'].astype(str).str.replace('.', '', regex=False).astype(int)
```

```python
df[['Votes']]
```

```python
df[['Votes']]
```

```python
df[['Score']]
```

```python
df['Score'] = pd.to_numeric(
    df['Score'].astype(str)
    .str.replace(r'[^0-9.,]', '', regex=True)
    .str.replace(',', '.')
    .str.replace(':', '.')
    .str.replace(r'\.+', '.', regex=True)
    .str.replace(r'^0+(?=\d)', '', regex=True)
    .str.strip('.'),
    errors='coerce'
)
```

```python
df['Score'] = df['Score'].astype(str).str.extract(r'(\d\.\d)')[0]
```

```python
df['Score'] = df['Score'].fillna(0)
```

```python
df['Score'] = df['Score'].astype(float)
```
